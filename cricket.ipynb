```python
import numpy as np
import pandas as pd
```

```python
cricket=pd.read_csv("all_matches.csv")
```

```python
cricket.columns
cricket.head()
cricket.head().to_csv("data_science.csv")
```

```python

ipl_data=cricket.drop(columns=['other_wicket_type','other_player_dismissed', 'season', 
                               'start_date','wicket_type', 'player_dismissed'])
ipl_data.columns
```

```python
ipl_data["total_runs"]=ipl_data["runs_off_bat"]+ipl_data['extras']
```

```python
ipl_data.drop(columns=['runs_off_bat','extras'])
ipl_data=ipl_data[ipl_data['ball']<5.7]
ipl_data=ipl_data[ipl_data['innings']<3]
```

```python
ipl_data=ipl_data.groupby(['match_id',
                  'venue',
                 'innings',
                 'batting_team',
                 'bowling_team']).total_runs.sum()
ipl_data.reset_index()
ipl_data=ipl_data.drop(columns=['match_id'])
ipl_data.to_csv("my_data.csv")
```
